```
MODEL_FOLDER = '/content/drive/My Drive/Models'
DETECTION_MODEL_URL = 'http://download.tensorflow.org/models/object_detection/faster_rcnn_inception_v2_coco_2018_01_28.tar.gz'
DETECTION_MODEL_TARGZ_PATH = os.path.join(MODEL_FOLDER, 'faster_rcnn_inception_v2_coco_2018_01_28.tar.gz')
DETECTION_MODEL_PATH = os.path.join(MODEL_FOLDER, 'faster_rcnn_inception_v2_coco_2018_01_28', 'frozen_inference_graph.pb')
DETECTION_MODEL_RELATIVE_PATH = os.path.join('faster_rcnn_inception_v2_coco_2018_01_28', 'frozen_inference_graph.pb')
```

```
VIDEOS_FOLDER = os.path.join('/tmp/', 'megapolis_football')
IMAGES_FOLDER  = os.path.join('/content/drive/My Drive/', 'megapolis_football', 'images')
INTERMEDIATE_FOLDER = os.path.join('/content/drive/My Drive/', 'megapolis_football', 'intermediate') 

YOUTUBE_PLAYLIST = 'https://www.youtube.com/playlist?list=PLyoTlBZWo2x91UFt9J8w7Lu3lniG4PxIg'
```

```
if not os.path.exists(MODEL_FOLDER):
    os.mkdir(MODEL_FOLDER)
if not os.path.exists(VIDEOS_FOLDER):
    os.mkdir(VIDEOS_FOLDER)
```

```
if not os.path.exists(DETECTION_MODEL_PATH):
  wget.download(DETECTION_MODEL_URL, DETECTION_MODEL_TARGZ_PATH)
  with tarfile.open(DETECTION_MODEL_TARGZ_PATH, 'r:gz') as tar_file:
    tar_file.extractall(MODEL_FOLDER, members=[m for m in tar_file.getmembers() if m.name == DETECTION_MODEL_RELATIVE_PATH])
```

```
processed_videos = [de.name.partition('_intermediate_')[0] 
                    for de in os.scandir(INTERMEDIATE_FOLDER) 
                    if de.name.partition('_intermediate_')[1] == '_intermediate_']
```

```
processed_videos
```

```

odapi = DetectorAPI(path_to_ckpt=DETECTION_MODEL_PATH)
settings = Settings( score_threshold = 0.4,
                     frame_capture_rate_sec = 5. ,
                     min_height_to_width_ratio = 0.0,
                     keep_intermediate_in_memory=True,
                     save_intermediate=False
                   )
personImagesExtractor = PersonImagesExtractor(odapi, settings) 
```

```
proxy_handler = {
"http": "78.37.94.172:8080",
'https': 'socks5://95.174.67.50:18080'
}
pl = Playlist(YOUTUBE_PLAYLIST, proxies=None)
```

```

youtube_links = [v for v in pl.video_urls]
for url in youtube_links:
    try:
        start = time.time()
        yt = YouTube(url)
        if yt.title in processed_videos:
            print(f'Video {yt.title} has already processed. Skipping.')
            continue
        print(f"Starting video {yt.title} at {time.ctime()}")
        def on_complete_callback(stream, file_handle):
            print(f'\nDownloading completed. File {file_handle} created.')
            return 

        yt.register_on_complete_callback(on_complete_callback)
        def on_progress_callback(chunk, file_handle, bytes_remaining):
            print(f'\rRemaining to download: {bytes_remaining>>10:9,} KiB out of {stream.filesize>>20:5,} MiB.  ', end='')
            return

        yt.register_on_progress_callback(on_progress_callback)
        stream = yt.streams.filter(only_video=True,file_extension='mp4').order_by('resolution').desc().first()

        if stream:
            out_file = stream.download(VIDEOS_FOLDER)
            finish = time.time()
            print(f"Video from {url} has been downloaded for {(finish-start)/60.:.1f} minutes")
            #personImagesExtractor.extract_all_images(out_file, IMAGES_FOLDER, save_images_on_the_fly = False, finish_at_sec=None)
            #path = personImagesExtractor.save_intermediate(intermediate_folder = INTERMEDIATE_FOLDER, file_prefix=out_file)
            #os.remove(out_file)
        else:
            print(f'Stream in {url} is not defined')
    except Exception as exc:
        print(f'Error while processing video from {url}:\n {str(exc)}')

```

```
videos_to_process = [de.path for de in os.scandir(VIDEOS_FOLDER) ]
for out_file in videos_to_process:
    print(f"Processing video {out_file} started at {time.ctime()}")
    personImagesExtractor.extract_all_images(out_file, IMAGES_FOLDER, save_images_on_the_fly = False, finish_at_sec=None)
    path = personImagesExtractor.save_intermediate(intermediate_folder = INTERMEDIATE_FOLDER, file_prefix=out_file)
    #os.remove(out_file) 
```
